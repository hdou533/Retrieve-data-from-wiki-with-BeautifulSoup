```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_banks'
wiki_data = requests.get(url)
```

```python
soup = BeautifulSoup(wiki_data.content, 'html.parser')
```

```python
# locate the table
table = soup.find_all('table')[3]
```

```python
print(table.prettify())
```

```python
# table rows
rows = table.find_all('tr')
```

```python
# table cells
td = rows[1].find_all('td')
```

```python
td
```

```python
#print(td[1].text)
```

```python
#print(td[2].text)
```

```python
# write table into a list
list = []
for row in rows:
    td = row.find_all('td')
    row = [row.text for row in td]
    list.append(row)
```

```python
list
```

```python
# list to a dataframe
df = pd.DataFrame(list, columns=['Rank','Bank','Market Cap'])
```

```python
df
```

```python
#cleaning part begins
df = df.drop(labels=0,axis=0)
df = df[['Bank','Market Cap']]
df
```

```python
df['Bank'] = df['Bank'].map(lambda x: x.rstrip('\n'))
df['Market Cap'] = df['Market Cap'].map(lambda x: x.rstrip('\n'))
```
